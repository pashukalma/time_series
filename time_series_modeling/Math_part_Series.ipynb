### Power and Taylor Series

### **Properties of Power Series**

We can add, subtract, multiply, divide, differentiate, and integrate convergent power series, just as we do with polynomials.

Consider a power series $\sum_{n=0}^{\infty} a_n x^n$ which converges for $|x| < A$.

**Differentiation:**

#### The derivative of the power series is obtained by differentiating term-by-term:
$$ \frac{d}{dx} \left( \sum_{n=0}^{\infty} a_n x^n \right) = \sum_{n=1}^{\infty} n a_n x^{n-1} $$
This derivative series also converges for $|x| < A$.

**Integration:**

#### The integral of the power series is obtained by integrating term-by-term:
$$ \int \left( \sum_{n=0}^{\infty} a_n x^n \right) dx = C + \sum_{n=0}^{\infty} \frac{a_n}{n+1} x^{n+1} $$
where $C$ is the constant of integration. This integral series also converges for $|x| < A$.

**Addition and Subtraction:**

#### Consider another power series $\sum_{n=0}^{\infty} b_n x^n$ which converges for $|x| < B$.
If $A \neq B$, then the sum (or difference) of the two power series is given by:
$$ \sum_{n=0}^{\infty} a_n x^n \pm \sum_{n=0}^{\infty} b_n x^n = \sum_{n=0}^{\infty} (a_n \pm b_n) x^n $$

#### This resulting series converges for $|x| < \min(A, B)$.
If $A = B$, then the sum (or difference) of the two power series is still:
$$ \sum_{n=0}^{\infty} a_n x^n \pm \sum_{n=0}^{\infty} b_n x^n = \sum_{n=0}^{\infty} (a_n \pm b_n) x^n $$

#### In this case, the radius of convergence of the resulting series is at least $A$, but it could be larger than $A$.\

```python
def term_series(a, n):
  """ Calculates the term of the series: (1 - a^(n+1))/(1-a) """
  if abs(1 - a) < 1e-9:  return n + 1
  else: return (1 - a**(n+1)) / (1 - a)

def calculate_SN(a, N):
  """ Calculates the partial sum S_N = Sum_{n=1}^{N} (1 - a^(n+1))/(1-a)
  Returns:
    A list of partial sums S_1, S_2, ..., S_N. """
  SN_values = []
  current_sum = 0
  for n in range(1, N + 1):
    current_sum += term_series(a, n)
    SN_values.append(current_sum)
  return SN_values
```

### l'Hopital's rule review

Evaluate limits with intermediate forms $\frac{0}{0}$, $\frac{\infty}{\infty}$

Rearrange limits of intermediate forms $0 \cdot \infty$, $\infty - \infty$ to the form $\frac{0}{0}$, $\frac{\infty}{\infty}$

Rearrange limits of exponential intermediate forms $0^0$, $1^\infty$, $\infty^0$ by applying $e^{\ln}$ to get a standard indeterminate form $\frac{0}{0}$, $\frac{\infty}{\infty}$

Apply l'Hopital's rule to compare the growth rates of functions at infinity

**l'Hopital Rule**

$f(x) \rightarrow \infty$
$g(x) \rightarrow \infty \quad \text{as } x \rightarrow a$
$f'/g' \rightarrow L$

$f/g \rightarrow L \text{ as } x \rightarrow a \qquad a = \pm \infty \text{ and } L = \pm \infty$

**Rate of Growth** ($x \rightarrow \infty$)

Consider $f, g>0$, we say $g$ grows faster than $f$ as $x$ goes to $\infty \quad 0 <f(x) \ll g(x)$

if $f,g \rightarrow \infty$ and $f(x)/g(x) \rightarrow 0$ as$ x \rightarrow \infty$

example: $\ln x \ll x^p \ll e^x \ll e^{x^2}  for p>0$
$e^{x^2} = e^{x^2}, not (e^x)^2 = e^{2x}$

**Rate of Decay** ($x \rightarrow \infty$)

Consider $f,g >0$, $f$ decays faster than  $g$ as $x$ tends toward $\infty \quad g(x) \gg f(x) \gg 0$

if $f,g \rightarrow 0 and f(x)/g(x) \rightarrow 0 as x \rightarrow \infty$
example: $1/\ln(x) \gg 1/x^p \gg e^{-x} for p>0$

### l'Hopital Rule examples

This limit is indeterminate as it is of the form $\frac{0}{0}$.
\begin{equation*}
\lim_{x \to 0^+} \frac{e^x - 1}{x - \sin(x)} = \lim_{x \to 0^+} \frac{\frac{d}{dx}(e^x - 1)}{\frac{d}{dx}(x - \sin(x))} = \lim_{x \to 0^+} \frac{e^x}{1 - \cos(x)} = \infty
\end{equation*}

This limit is of the form $\frac{\infty}{\infty}$, so we apply l'Hôpital's Rule.
\begin{equation*}
\lim_{x \to \infty} \frac{\ln(x^2) + 4x + 30}{8x + \sqrt{x}} = \lim_{x \to \infty} \frac{\frac{d}{dx}(\ln(x^2) + 4x + 30)}{\frac{d}{dx}(8x + \sqrt{x})} = \lim_{x \to \infty} \frac{\frac{2}{x} + 4}{8 + \frac{1}{2\sqrt{x}}} = \frac{4}{8} = \frac{1}{2}
\end{equation*}

This limit is of the form $\frac{0}{2}$, so l'Hôpital's rule does not apply. The limit is zero.
\begin{equation*}
\lim_{x \to 0} \frac{x \sin(x)}{\cos^2(x) + 1} = 0
\end{equation*}

This limit is of the form $\frac{\infty}{\infty}$, so we apply l'Hôpital's rule.
\begin{align*}
\lim_{x \to \infty} \frac{\ln(e^x + 5x)}{4x + 3} &= \lim_{x \to \infty} \frac{\frac{d}{dx}(\ln(e^x + 5x))}{\frac{d}{dx}(4x + 3)} \\
&= \lim_{x \to \infty} \frac{\frac{e^x + 5}{e^x + 5x}}{4} \\
&= \lim_{x \to \infty} \frac{1}{4} \frac{e^x + 5}{e^x + 5x} \\
&= \frac{1}{4} \lim_{x \to \infty} \frac{e^x}{e^x} \\
&= \frac{1}{4}
\end{align*}

This limit is an intermediate form of the type $\frac{\infty}{\infty}$,

applying L'Hôpital's rule $\frac{x^{10000}}{e^x} \sim \frac{10000x^{9999}}{e^x} \sim  \frac{10000 \cdot 9999 \cdot x^{9998}}{e^x} \sim \dots \sim \frac{10^5}{e^x} \to 0 $

$\frac{x^n}{e^x} \overset{L'H}{ \sim} \dots \sim \frac{n!}{e^x} \to 0 \text{ for } x \to \infty $
$e^x$ grows faster than $x^n$ for any $n$. $e^x$ grows faster than any polynomial.

### Indeterminate forms involving exponents

**Case 1:**

$1^{\infty}$ form
Given $\lim_{x \to a} f(x) = 1$ and $\lim_{x \to a} g(x) = +\infty$, find $\lim_{x \to a} [f(x)]^{g(x)}$.

We can consider plugging in some numbers like $(1.0001)^{(10^6)}$ or $(0.9995)^{(10^6)}$ to get an intuition.

The approach is to take the logarithm of the function:

$ \ln([f(x)]^{g(x)}) = g(x) \ln(f(x)) = \frac{\ln(f(x))}{\frac{1}{g(x)}} $
As $x \to a$, we have $\ln(f(x)) \to \ln(1) = 0$ and $\frac{1}{g(x)} \to \frac{1}{+\infty} = 0$. Thus, the expression becomes of the form $\frac{0}{0}$, which is an indeterminate form.

Therefore, the original function can be written as:

$ [f(x)]^{g(x)} = e^{\ln([f(x)]^{g(x)})} = e^{\frac{\ln(f(x))}{\frac{1}{g(x)}}} $
To find the limit, we need to evaluate the limit of the exponent $\frac{\ln(f(x))}{\frac{1}{g(x)}}$ as $x \to a$ and then exponentiate the result.



### Exponents continued
**Case 2**:

$0^{\infty}$ form
Given $f(x) \geq 0$, $\lim_{x \to a} f(x) = 0$, and $\lim_{x \to a} g(x) = +\infty$, find $\lim_{x \to a} [f(x)]^{g(x)}$.

Plugging in some numbers like $(10^{-10})^{(10^{10})}$ suggests that the limit is close to zero.
Intuitively, a number close to zero raised to a large power becomes even closer to zero.

Let's examine the logarithm of the function:
$$ \ln([f(x)]^{g(x)}) = g(x) \ln(f(x)) $$
As $x \to a$, we have $\ln(f(x)) \to \ln(0) = -\infty$ and $g(x) \to +\infty$. Thus, the expression becomes of the form $(+\infty) \cdot (-\infty)$, which tends to $-\infty$.

Therefore, the original function tends to 0 as $x$ approaches $a$:
$$ \lim_{x \to a} [f(x)]^{g(x)} = \lim_{x \to a} e^{\ln([f(x)]^{g(x)})} = \lim_{x \to a} e^{g(x) \ln(f(x))} = e^{-\infty} = 0 $$

### Exponents continued
**Case 3**:

$\infty^{0}$ form
Given $\lim_{x \to a} f(x) = +\infty$ and $\lim_{x \to a} g(x) = 0$, the limit $\lim_{x \to a} [f(x)]^{g(x)}$ is of indeterminate form $\infty^{0}$.


**Case 4**:

$0^{0}$ form
Given $\lim_{x \to a} f(x) = 0$ and $\lim_{x \to a} g(x) = 0$, the limit $\lim_{x \to a} [f(x)]^{g(x)}$ is of indeterminate form $0^{0}$.

### Singularity function

A singularity function $f(x)$ is a point $x=s$ such that $f(x)$ does not exist (DNE) at $x=s$. There are three main ways that the function can fail to exist at a point:

- $\lim_{x \to s^+} |f(x)|$ and/or $\lim_{x \to s^-} |f(x)|$ tends to $\infty$.
- $\lim_{x \to s^+} f(x)$ DNE. $f(x)$ may oscillate or have a jump discontinuity.
- $\lim_{x \to s^\pm} f(x)$ exists and is finite, $f(x)$ has a removable discontinuity.

**Examples**

- $\frac{1}{\sqrt{x^2 + 5}}$ - has no singularities.
- $\frac{1}{x^2 + x}$ - has singularities at $x = 0$, and $x = -1$.
- $\frac{e^{-x}}{x}$ - has a singularity at $x = 0$.
- $\frac{\ln|x|}{x^2 - 1}$ - has singularities at $x = 1$, $x = 0$, and $x = -1$.
  - At $x = 0$, the function goes to $\infty$.
    - We evaluate the limit at $x = \pm 1$:
    - $x \to 1$: $\frac{\ln|x|}{x^2 - 1} = \frac{\ln(x)}{x^2 - 1} \sim \frac{1/x}{2x} \to \frac{1}{2}$.
    - $x \to -1$: $\frac{\ln|x|}{x^2 - 1} = \frac{\ln(-x)}{x^2 - 1} \sim \frac{1/x}{2x} \to \frac{1}{2}$.
    
    These points $x = \pm 1$ are removable discontinuities of this function.

**Integrals**
- $\int_1^\infty \frac{1}{\sqrt{x^2 + 5}} \, dx$ - has no singularities.
- $\int \frac{1}{x^2 + x} \, dx$ - has singularities at $x = 0$, and $x = -1$, but the integral is over an interval that doesn't contain these points.
- $\int \frac{e^{-x}}{x} \, dx$ - has a singularity at $x = 0$.
- $\int \frac{\ln(x)}{x^2} \, dx$ - has a singularity at $x = 0$, but at this point is not in the interval we are integrating over.

### Alternating series:

An alternating series is a series whose terms alternate in signs.
\begin{equation*}
    \pm \sum_{n=1}^{\infty} (-1)^n c_n, \quad \text{where } c_n \geq 0
\end{equation*}
Test for convergence:

if for all $n$ large enough, $c_n$ decreases as $n$ increases, then $\pm \sum_{n=1}^{\infty} (-1)^n c_n$ converges.


**Examples of alternate series**

$\sum x^n$ geometric series converges in $|x| < 1$, diverges on $|x| \geq 1$.

$\sum \frac{1}{n^p}$ p-series converges in $p > 1$, diverges on $p \leq 1$.

$\sum \frac{1}{n (\ln(n)^p)}$ p-series converges in $p > 1$, diverges on $p \leq 1$.

$\sum \frac{\ln(n)}{n^p}$ p-series converges in $p > 1$, diverges on $p \leq 1$.

$\sum \frac{x^n}{n!}$ converges for all $x$.

$\sum \frac{x^n}{n^p}$ converges for $|x| < 1$, diverges for $|x| > 1$.

$\sum \frac{n!}{n^n}$ converges.
